```python
data = df["Passengers"].values
```

```python
data[:5]
```

```python
type(data[:5])
```

* Biz pandas dataframe'i burada numpy tipine dönüştürdük. Çünkü RNN'de bu şekilde kullanmak istiyoruz. RNN zaman serileri üzerine işlemi gerçekleştirebilmektedir.

```python
type(data[0])
```

```python
data.shape
```

```python
data = data.reshape(-1, 1)
```

```python
data.shape
```

```python
# Şimdi verimizi train-test olarak ikiye ayıracağız 
def split_data(dataframe, test_size):
    position = int(round(len(dataframe) * (1 - test_size)))
    train = dataframe[: position]
    test = dataframe[position : ]
    return train, test, position
```

```python
train, test, position = split_data(data, 0.33)  # Burada %33 olma nedeni verimizin tam bölünebilmesi içindir
```

```python
print(train.shape, test.shape)
```

* Gradient Descent temelli yöntemlerde standartlaştırma işlemleri yapılır. Bunun yapılma nedeni optimizasyon süresini daha hızlı bir şekilde gerçekleştirmek yani hız ve performans kazandırmaktır. Eğer "Gradient Descent" temelli bir yöntem kullanılıyorsa değişkenleri 0 ile 1 arasında dönüştürmek lazım. 

```python
scaler_train = MinMaxScaler(feature_range = (0, 1))
```

```python
train = scaler_train.fit_transform(train)
```

```python
scaler_test = MinMaxScaler(feature_range = (0, 1))
```

```python
test = scaler_test.fit_transform(test)
```

```python
train[0:5]
```

```python
test[0:5]
```

```python
def create_features(data, lookback):
    X, Y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i - lookback:i, 0])
        Y.append(data[i, 0])

    return np.array(X), np.array(Y)
```

```python
lookback = 1
```

```python
# Train Veri Seti
X_train, y_train = create_features(train, lookback)
```

```python
# Test Veri Seti
X_test, y_test = create_features(test, lookback)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

* Yukarıda yeniden boyutlandırma işlemi RNN için oluşturulmuş uygun formatın sağlanması için yapılmaktadır.

# 4) Modelleme (Modeling)

```python
model = Sequential()

model.add(SimpleRNN(units = 50, # Nöron sayısını ifade etmektedir. Biz 50 adet olarak ayarlamış bulunmaktayız. 
                    # Bu nöronlar kendi aralarında veri transferleri gerçekleştirebilimektedirler.
                   activation = "relu",
                   input_shape = (X_train.shape[1], lookback)))

model.add(Dropout(0.2))
model.add(Dense(1))
```

```python
model.summary()
```

### Optimizasyon ve Model Değerlendirme Metrikleri Ayarlanması

```python
model.compile(loss = "mean_squared_error", optimizer = "adam")
```

```python
callbacks = EarlyStopping(monitor='val_loss', patience=5, verbose = 1, restore_best_weights=True)
```

```python
history = model.fit(x = X_train,
                    y = y_train,
                    epochs=50,
                    batch_size = 1,
                    validation_data = (X_test, y_test),
                    callbacks = callbacks,
                    shuffle = False)
```

```python
#--------------
# Loss Grafiği
#--------------
plt.figure(figsize = (15,6))
plt.subplot(1, 2, 2)
plt.plot(history.history["loss"], color = "blue", label = "Training Loss")
plt.plot(history.history["val_loss"], color = "red", label = "Validation Loss")
plt.legend(loc = "upper right")
plt.xlabel("Epoch", fontsize = 16)
plt.ylabel("Loss", fontsize = 16)
plt.ylim([0, max(plt.ylim())])
plt.title("Training and Validation Loss", fontsize = 16)
plt.show()
```

# 5) Değerlendirme (Evolation)

```python
loss = model.evaluate(X_test, y_test, batch_size = 1)
print("\nTest Loss: %.1f.%%" % (100.0 * loss))
```

```python
train_predict = model.predict(X_train)
```

```python
test_predict = model.predict(X_test)
```

```python
train_predict = scaler_train.inverse_transform(train_predict)
test_predict = scaler_test.inverse_transform(test_predict)
```

```python
y_train = scaler_train.inverse_transform(y_train)
y_test = scaler_test.inverse_transform(y_test)
```

```python
# Train veri setine ait RMSE değeri
train_rmse = np.sqrt(mean_squared_error(y_train, train_predict))
```

```python
# Test veri setine ait RMSE değeri
test_rmse = np.sqrt(mean_squared_error(y_test, test_predict))
```

```python
print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
```

```python
test_prediction_df = result_df[position + lookback:]
test_prediction_df["Predicted"] = test_predict
test_prediction_df.head()
```
