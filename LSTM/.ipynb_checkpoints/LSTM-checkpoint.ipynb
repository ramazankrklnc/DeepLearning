# 1) İş Problemi (Business Problem)

* **Tesla şirketine ait gelecek dönemlerdeki hisse senedi kapanış fiyatlarını tahmin etmeye çalışmaktır.**

-------------------------------------------------------------------------------------------------------------
* **Date:** İlgili İşlem Gününün tarihini temsil etmektedir.
* **Open:** İlgili işlem gününe ait Hisse başlangıç fiyatını temsil etmektedir.
* **High:** İlgili işlem gününe ait en yüksek fiyatı temsil etmektedir.
* **Low:** İlgili işlem gününe ait en düşük fiyatı temsil etmektedir.
* **Close:** İlgili işlem gününde hisse senedinin kapanış fiyatını temsil etmektedir.
* **Adj Close:** İlgili işlem gününe ait hisse senedinin düzeltilmiş kapanış fiyatını temsil etmektedir.
* **Volume:** İlgili işlem gününe ait işlem hacmi bilgisini temsil etmektedir.

# 2) Veriyi Anlamak (Data Understanding)

```python
# Veri okuma ve işleme adımında kullanılacak kütüphaneler 
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt

# Model değerlendirme ve veriyi scale edebilmemmiz için kullanılacak olan kütüphaneler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

#  Model için kullanılacak olan kütüphaneler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# Warningleri mesajlarını kapatmaya yaramaktadır.
from warnings import filterwarnings
filterwarnings("ignore")

# Tensorflow warning engelleme
import os 
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

```python
df = pd.read_csv("TSLA.csv")
df.head()
```

```python
def check_df(dataframe, head=5):
    print("##################### Shape #####################")
    print(dataframe.shape)
    print("##################### Types #####################")
    print(dataframe.dtypes)
    print("##################### Head #####################")
    print(dataframe.head(head))
    print("##################### Tail #####################")
    print(dataframe.tail(head))
    print("##################### NA #####################")
    print(dataframe.isnull().sum())
    print("##################### Quantiles #####################")
    print(dataframe.select_dtypes(include=['number']).quantile([0, 0.05, 0.50, 0.95, 0.99, 1]).T)
```

```python
check_df(df)
```

# 3) Veriyi Hazırlamak (Data Prepatation)

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
tesla_df = df[["Date", "Close"]]
```

```python
tesla_df.head()
```

```python
print("Minimum Tarih: ", tesla_df["Date"].min())
print("Maksimum Tarih: ", tesla_df["Date"].max())
```

```python
tesla_df.index = tesla_df["Date"]
```

```python
tesla_df.head()
```

```python
tesla_df.drop(["Date"], axis = 1, inplace = True)
```

```python
tesla_df.head()
```

```python
result_df = tesla_df.copy()
```

```python
plt.figure(figsize = (15,6))
plt.plot(tesla_df["Close"], color = "blue")
plt.title("Tesla Stock Price")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.show()
```

```python
tesla_df = tesla_df.values
```

```python
tesla_df[0:5]
```

```python
tesla_df = tesla_df.astype("float32")
```

```python
# Şimdi verimizi train-test olarak ikiye ayıracağız 
def split_data(dataframe, test_size):
    pos = int(round(len(dataframe) * (1 - test_size)))
    train = dataframe[: pos]
    test = dataframe[pos: ]
    return train, test, pos
```

```python
train, test, position = split_data(tesla_df, 0.20)
```

```python
print(train.shape, test.shape)
```

```python
scaler_train = MinMaxScaler(feature_range = (0, 1))
```

```python
train = scaler_train.fit_transform(train)
```

```python
scaler_test = MinMaxScaler(feature_range = (0, 1))
```

```python
test = scaler_test.fit_transform(test)
```

```python
train[0:5]
```

```python
test[0:5]
```

```python
def create_features(data, lookback):
    X, Y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i - lookback:i, 0])
        Y.append(data[i, 0])

    return np.array(X), np.array(Y)
```

```python
lookback = 20
```

```python
# Train Veri Seti
X_train, y_train = create_features(train, lookback)
```

```python
# Test Veri Seti
X_test, y_test = create_features(test, lookback)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
X_train[:5]
```

```python
y_train[:5]
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

# 4) Modelleme (Modeling)

```python
model = Sequential()

model.add(LSTM(units = 50,
                   activation = "relu",
                   input_shape = (X_train.shape[1], lookback)))
model.add(Dropout(0.2))
model.add(Dense(1))
```

```python
model.summary()
```

### Optimizasyon ve Model Değerlendirme Metrikleri Ayarlanması

```python
model.compile(loss = "mean_squared_error", optimizer = "adam")
```

```python
callbacks = EarlyStopping(monitor='val_loss', patience = 3 , verbose = 1, restore_best_weights=True)
```

```python
history = model.fit(x = X_train,
                    y = y_train,
                    epochs = 100,
                    batch_size = 20,
                    validation_data = (X_test, y_test),
                    callbacks = callbacks,
                    shuffle = False)
```

```python
# Loss Grafiği
plt.figure(figsize = (15,6))
plt.subplot(1, 2, 2)
plt.plot(history.history["loss"], label = "Training Loss")
plt.plot(history.history["val_loss"], label = "Validation Loss")
plt.legend(loc = "upper right")
plt.xlabel("Epoch", fontsize = 16)
plt.ylabel("Loss", fontsize = 16)
plt.ylim([0, max(plt.ylim())])
plt.title("Training and Validation Loss", fontsize = 16)
plt.show()
```

# 5) Değerlendirme (Evolation)

```python
loss = model.evaluate(X_test, y_test, batch_size = 20)
print("\nTest Loss: %.1f.%%" % (100.0 * loss))
```

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
train_predict = scaler_train.inverse_transform(train_predict)
test_predict = scaler_test.inverse_transform(test_predict)
```

```python
y_train = scaler_train.inverse_transform(y_train)
y_test = scaler_test.inverse_transform(y_test)
```

```python
# Train veri setine ait RMSE değeri
train_rmse = np.sqrt(mean_squared_error(y_train, train_predict))
```

```python
# Test veri setine ait RMSE değeri
test_rmse = np.sqrt(mean_squared_error(y_test, test_predict))
```

```python
print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
```

```python
train_prediction_df = result_df[lookback:position]
train_prediction_df["Predicted"] = train_predict
train_prediction_df.head()
```

```python
test_prediction_df = result_df[position + lookback:]
test_prediction_df["Predicted"] = test_predict
test_prediction_df.head()
```

```python
# Train ve Test Tahminlerini ayrı ayrı görselleştirme işlemi

plt.figure(figsize = (15,6))
plt.plot(result_df, label = "Gerçek Değerler")
plt.plot(train_prediction_df["Predicted"], color = "blue", label = "Train Tahmini")
plt.plot(test_prediction_df["Predicted"], color = "red", label = "Test Tahmini")
plt.xlabel("Zaman")
plt.ylabel("Stok Sayısı")
plt.legend()
plt.show()
```
